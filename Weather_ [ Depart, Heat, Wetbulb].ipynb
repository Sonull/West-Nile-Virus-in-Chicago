```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('/Users/leeyeji/Desktop/Kaggle WNV/predict-west-nile-virus/weather.csv')
```

```python
#use this code only if you want to change the missing value "M" with "NaN" 
weather=weather.replace(['M'], np.NaN)
```

```python
weather.isnull().sum()
```

```python
#Split the dataset into station1 and station2
```

```python
station1 = weather.loc[weather['Station'] == 1] 
```

```python
station2 = weather.loc[weather['Station'] == 2] 
```

```python
#join the separated dataset 
```

```python
final_station = station1.join(station2.set_index('Date'), on='Date', lsuffix = '_Station1', rsuffix='_Station2')
```

---

```python
#Column name : Depart - since all missing values belong to Station2, 
#We have deicede to replace the missing value of station2 with the value of station1 where has the same date 
```

```python
#Replace the missing values of Depart_Station2 with values of Depart_Station1 
 # Use this code only if you changed the "M" to "NaN"
for col in ['Depart_Station2']:
    final_station[col] = final_station[col].fillna(final_station['Depart_Station1'])
```

```python
final_station[['Depart_Station1','Depart_Station2']]
```

```python
#Replace the missing values of Depart_Station2 with values of Depart_Station1 
 #If you still keep the missing value as "M" use this code 
final_station['Depart_Station2'] = np.where((final_station.Depart_Station2 == 'M'),
                                            final_station.Depart_Station1, final_station.Depart_Station2)
```

```python
final_station[['Depart_Station1','Depart_Station2']]
```

```python
final_station.isnull().sum()
```

---

```python
#Column:Heat - same as 'Depart' Column, all missing values belong to Station2, 
#We have deicede to replace the missing value of station2 with the value of station1 where has the same date 
```

```python
#Replace the missing values of Heat_Station2 with values of Heat_Station1 
 # Use this code only if you changed the "M" to "NaN"
for col in ['Heat_Station2']:
    final_station[col] = final_station[col].fillna(final_station['Heat_Station1'])
```

```python
#Replace the missing values of Heat_Station2 with values of Heat_Station1 
 #If you still keep the missing value as "M" use this code 
final_station['Heat_Station2'] = np.where((final_station.Heat_Station2 == 'M'),
                                            final_station.Heat_Station1, final_station.Heat_Station2)
```

```python
final_station[['Heat_Station1','Heat_Station2']]
```

```python
final_station.isnull().sum()
```

```python
#Replace M in Tavg with Tmax-Tmin / 2 + Tmin
weather['Tavg'].replace(to_replace='M',value = (weather['Tmax']-weather['Tmin'])/2 + weather['Tmin'], inplace=True)

#Change Tavg type from object to int
weather[["Tavg"]] = weather[["Tavg"]].astype(int)
```

---

```python
#Colum : Wetbulb - Since there are only 4 missing variables (station 1 : 3 missing variables, station2: 1 missing variables), 
#we are going toreplace missing values with the mean of each station)
```

```python
#change the type of Wetbulb_Station1 Column to numeric 
final_station['WetBulb_Station1']= pd.to_numeric(final_station['WetBulb_Station1'], errors = 'coerce')
```

```python
# calculate mean of WetBulb_Station1 = WetBulb_Station1_mean
WetBulb_Station1_mean = final_station['WetBulb_Station1'].mean(skipna = True)
```

```python
WetBulb_Station1_mean
```

```python
#replace the missing value of WetBulb_Station1 with the WetBulb_Station1_mean
final_station['WetBulb_Station1'] = final_station.WetBulb_Station1.fillna(WetBulb_Station1_mean)
```

```python
#change the type of Wetbulb_Station2 Column to numeric 
final_station['WetBulb_Station2']= pd.to_numeric(final_station['WetBulb_Station2'], errors = 'coerce')
```

```python
# calculate mean of WetBulb_Station2
WetBulb_Station2_mean = final_station['WetBulb_Station2'].mean(skipna = True)
```

```python
WetBulb_Station2_mean
```

```python
#replace the missing value of WetBulb_Station2 with the WetBulb_Station2_mean
final_station['WetBulb_Station2'] = final_station.WetBulb_Station2.fillna(WetBulb_Station2_mean)
```

```python
def to_celcius(x):
    c = ((x-32)/9)*5
    return(c)
```

```python
def rel_hum(dry,wet,press=0.6687451584):
    e = float(math.e)
    ed = 6.112*(e**((17.502*dry)/(240.97+dry)))
    ew = 6.112*(e**((17.502*wet)/(240.97+wet)))
    result = (ew-press*(1+.00115*wet)*(dry-wet))/ed*100
    return(result)
```

```python
final_station['Tavg_Station1']= pd.to_numeric(final_station['Tavg_Station1'], errors = 'coerce')
final_station['Tavg_Station2']= pd.to_numeric(final_station['Tavg_Station2'], errors = 'coerce')
#final_station['StnPressure_Station1']= pd.to_numeric(final_station['StnPressure_Station1'], errors = 'coerce')
```

```python
import math
final_station['WetBulb_Station1_c']=final_station.WetBulb_Station1.apply(to_celcius)
final_station['Tavg_Station1_c']=final_station.Tavg_Station1.apply(to_celcius)

final_station['WetBulb_Station2_c']=final_station.WetBulb_Station2.apply(to_celcius)
final_station['Tavg_Station2_c']=final_station.Tavg_Station2.apply(to_celcius)

final_station['rel_hum_station1']=rel_hum(final_station['Tavg_Station1_c'],final_station['WetBulb_Station1_c'])
final_station['rel_hum_station2']=rel_hum(final_station['Tavg_Station2_c'],final_station['WetBulb_Station2_c'])
```

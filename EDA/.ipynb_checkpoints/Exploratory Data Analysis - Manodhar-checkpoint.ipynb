```python
import pandas as pd
master_database=pd.read_csv("train_spray_weather_v1.csv",float_precision='round_trip')
master_database['yr_mo']=master_database['year'].astype(str) + '-' + master_database['month'].astype(str)
```

```python
#WNV detected traps by year
db1 = master_database[master_database["WnvPresent"]==1]
EDA1=db1.groupby('yr_mo').nunique()['Trap'].reset_index()
```

```python
db2 = master_database[master_database["WnvPresent"]==0]
EDA2=db2.groupby('yr_mo').nunique()['Trap'].reset_index()
```

```python
EDA3=EDA2.merge(EDA1, on='yr_mo',suffixes=('_no_WNV', '_WNV'),how='left')
```

```python
EDA3.plot.bar(x='yr_mo',figsize=(15,10))
```

```python
EDA4=db1.groupby('yr_mo')['Tmin_Station','Tmax_Station'].mean().reset_index()
print(EDA4)
```

```python
#Temp trend with WNV virus
EDA4.plot.line(x='yr_mo',figsize=(15,10))
```

```python
EDA5=db1.groupby('Tavg_Station').nunique()['Trap'].reset_index()
EDA6=master_database.groupby('Tavg_Station').nunique()['Trap'].reset_index()
EDA7=EDA6.merge(EDA5, on='Tavg_Station',suffixes=('_tot', '_WNV'))
EDA7['per_trap_wnv']=EDA7['Trap_WNV']*100/EDA7['Trap_tot']

import matplotlib.pyplot as plt
import numpy as np
x = EDA7['Tavg_Station']
y = EDA7['per_trap_wnv']
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
EDA7
```

```python
EDA8=db1.groupby('rel_hum_station').nunique()['Trap'].reset_index()
EDA9=master_database.groupby('rel_hum_station').nunique()['Trap'].reset_index()
EDA10=EDA9.merge(EDA8, on='rel_hum_station',suffixes=('_tot', '_WNV'))
EDA10['per_trap_wnv']=EDA10['Trap_WNV']*100/EDA10['Trap_tot']

x = EDA10['rel_hum_station']
y = EDA10['per_trap_wnv']
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

```python
EDA14=db1.groupby('month').nunique()['Trap'].reset_index()
EDA15=master_database.groupby('month').nunique()['Trap'].reset_index()
EDA16=EDA15.merge(EDA14, on='month',suffixes=('_tot', '_WNV'))
EDA16['per_trap_wnv']=EDA16['Trap_WNV']*100/EDA16['Trap_tot']
EDA16.plot.bar(x='month', y='per_trap_wnv', rot=0)
```

```python
# Calculate min Spray distance
#Filter data for 2011 and 2013 as spray data is available in these 2 years 
EDA17=master_database#[(master_database.year == 2011) | (master_database.year == 2013)]
EDA17['min_spary_dist']=EDA17.iloc[:,15:25].min(axis=1)

#Filter for WNV traps
EDA18=db1#[(db1.year == 2011) | (db1.year == 2013)]
EDA18['min_spary_dist']=EDA18.iloc[:,15:25].min(axis=1)
```

```python
EDA17.min_spary_dist = EDA17.min_spary_dist.round() #master
EDA18.min_spary_dist = EDA18.min_spary_dist.round() #WNV
```

```python
EDA19=EDA18.groupby('min_spary_dist').nunique()['Trap'].reset_index()
EDA20=EDA17.groupby('min_spary_dist').nunique()['Trap'].reset_index()
EDA21=EDA20.merge(EDA19, on='min_spary_dist',suffixes=('_tot', '_WNV'))
EDA21['per_trap_wnv']=EDA21['Trap_WNV']*100/EDA21['Trap_tot']

x = EDA21['min_spary_dist']
y = EDA21['per_trap_wnv']
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
EDA21
```
